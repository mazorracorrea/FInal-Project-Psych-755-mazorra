**SENTIMENT ANALYSYS**

The first art of the code is to load the data set and aplly the VADER library for sentiment analysys.
Note: the dataset we must apply here if after we made the BERT classifier on the dataset of tweets.

**Structure:**

Setup: Installs the vaderSentiment library and other libraries.

Data: The main classified dataset (output from the BERT+LGBM model) is loaded into a pandas DF.

VADER: An instance of the SentimentIntensityAnalyzer is created.

Sentiment Score Calculation: A function get_vader_scores is defined to apply the polarity_scores method to the sanitized text. This method returns negative, neutral, positive, and compound score. This function is applied to the santext column of the entire dataset. A progress bar from tqdm is used to track the process (we include this column for this process because in the first iteration of the project the kernell crashed and we never benn sure on how the process was going).

Output: The sentiment scores are added as new columns to the DF: vader_neg, vader_neu, vader_pos, and vader_compound.

The final DF is saved to a new CSV file, debate_tweets_with_polarization_and_sentiment.csv, in a processed data folder.



```python
#The first art of the code is to load the data set and aplly the VADER library for sentiment analysys.
#Note: the dataset we must apply here if after we made the BERT classifier on the dataset of tweets.

import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tqdm import tqdm
import os


# Load classified tweet dataset
df = pd.read_csv("debate_tweets_classified_COMPLETE.csv")

# Initialize VADER as an analyzer for the tweets
analyzer = SentimentIntensityAnalyzer()

# Define function to get compound score
def get_vader_scores(text):
    if isinstance(text, str):
        return analyzer.polarity_scores(text)
    else:
        return {'neg': 0.0, 'neu': 0.0, 'pos': 0.0, 'compound': 0.0}

# Apply VADER
print("Calculating VADER sentiment scores...")
tqdm.pandas()
df[['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound']] = df['santext'].progress_apply(
    lambda x: pd.Series(get_vader_scores(x))
)


# Save the updated dataset in a new directory
os.makedirs("data/processed", exist_ok=True)
df.to_csv("data/processed/debate_tweets_with_polarization_and_sentiment.csv", index=False)
print("‚úÖ Sentiment scores added and file saved.")
```
